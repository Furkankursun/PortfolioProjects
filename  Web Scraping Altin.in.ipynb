```python
from bs4 import BeautifulSoup 
import requests
import smtplib
import time
import datetime
from datetime import datetime
import pandas as pd 
import csv
```

```python
# connet to website 

Url='https://altin.in/'

head={"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36"}


page= requests.get(Url,headers=head)       

Soup1=BeautifulSoup(page.content,"html.parser")

Soup2=BeautifulSoup(Soup1.prettify(),"html.parser")


dolar=Soup2.find(id="dfiy").get_text()
dolar=dolar.strip()

euro=Soup2.find(id="efiy").get_text()
euro=euro.strip()

sterlin=Soup2.find(id="sfiy").get_text()
sterlin=sterlin.strip()


now = datetime.now()
dt_now = now.strftime("%d/%m/%Y %H:%M:%S")

print('1 $ =',dolar,'₺')

print('1 € =',euro,'₺')

print('1 £ =',sterlin,'₺')
```

```python
header = ['Dolar', 'Euro','Sterlin','Date']
data = [dolar, euro,sterlin, dt_now]


with open('AltıninScraperDataset.csv', 'w', newline='', encoding='UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerow(data)
```
